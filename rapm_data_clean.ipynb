```python
import pandas as pd
import numpy as np


merged_df = pd.read_csv("merged_2016_2017_2018.csv").reset_index(drop=True)
merged_df = merged_df.loc[:, ~merged_df.columns.str.contains('^Unnamed')]
merged_df.head()
```

```python
winners_df = pd.read_csv("merged_winners.csv").reset_index(drop=True)
winners_df = winners_df.loc[:, ~winners_df.columns.str.contains('^Unnamed')]
winners_df["HOME_TEAM_WON"] = (winners_df["WINNER_SIDE"] == "HOME").astype(int)

winners_df.head(20)
```

```python
import unicodedata

# Function to remove accents from text
def remove_accents(text):
    if pd.isna(text):
        return text
    # Normalize to NFD (decomposed form) and filter out combining characters
    nfd = unicodedata.normalize('NFD', text)
    return ''.join(char for char in nfd if unicodedata.category(char) != 'Mn')

# Clean player names in final_df
merged_df['PLAYER_NAME_CLEAN'] = merged_df['PLAYER_NAME'].apply(remove_accents)
```

```python
rapm_df_2016 = pd.read_csv("rapm_2016-17.csv").reset_index(drop=True)
rapm_df_2017 = pd.read_csv("rapm_2016-17.csv").reset_index(drop=True)
rapm_df_2018 = pd.read_csv("rapm_2016-17.csv").reset_index(drop=True)
```

```python
merged_df_2016 = merged_df[merged_df["season"] == "2016-17"]
merged_df_2017 = merged_df[merged_df["season"] == "2017-18"]
merged_df_2018 = merged_df[merged_df["season"] == "2018-19"]
```

```python
# clean rapm dataset

rapm_df_2016_clean = (
    rapm_df_2016
    .sort_values("Poss", ascending=False)
    .drop_duplicates(subset="Player", keep="first")
)

rapm_df_2017_clean = (
    rapm_df_2017
    .sort_values("Poss", ascending=False)
    .drop_duplicates(subset="Player", keep="first")
)

rapm_df_2018_clean = (
    rapm_df_2018
    .sort_values("Poss", ascending=False)
    .drop_duplicates(subset="Player", keep="first")
)
```

```python
merged_df_2017.shape
```

```python
players_mdf_2016 = pd.merge(
    merged_df_2016,
    rapm_df_2016_clean,
    left_on="PLAYER_NAME_CLEAN",
    right_on="Player",
    how="left"
)

players_mdf_2017 = pd.merge(
    merged_df_2017,
    rapm_df_2017_clean,
    left_on="PLAYER_NAME_CLEAN",
    right_on="Player",
    how="left"
)

players_mdf_2018 = pd.merge(
    merged_df_2018,
    rapm_df_2018_clean,
    left_on="PLAYER_NAME_CLEAN",
    right_on="Player",
    how="left"
)

players_mdf_2018.shape
```

```python
# Roughly 25% missing values

players_mdf_2016_filled = players_mdf_2016.fillna(players_mdf_2016.mean(numeric_only=True))
players_mdf_2017_filled = players_mdf_2017.fillna(players_mdf_2017.mean(numeric_only=True))
players_mdf_2018_filled = players_mdf_2018.fillna(players_mdf_2018.mean(numeric_only=True))

#players_mdf_2018[players_mdf_2018["Team"].isna()]
```

```python
def summarize_games(df):
    num_cols = ["ORAPM", "DRAPM", "RAPM", "Rank"]
    df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

    result_rows = []
    for game_id, group in df.groupby("GAME_ID"):
        teams = group["TEAM_ABBREVIATION"].unique()
        if len(teams) != 2:
            continue

        team1, team2 = teams[0], teams[1]
        t1 = group[group["TEAM_ABBREVIATION"] == team1]
        t2 = group[group["TEAM_ABBREVIATION"] == team2]

        row = {
            "GameID": game_id,
            "Team1": team1,
            "Team2": team2,
            "AVG_ORAPM_Team1": t1["ORAPM"].mean(),
            "AVG_DRAPM_Team1": t1["DRAPM"].mean(),
            "AVG_ORAPM_Team2": t2["ORAPM"].mean(),
            "AVG_DRAPM_Team2": t2["DRAPM"].mean(),
            "AVG_RANK_Team1": t1["Rank"].mean(),
            "AVG_RANK_Team2": t2["Rank"].mean()
        }

        result_rows.append(row)

    return pd.DataFrame(result_rows)
```

```python
games_2016 = summarize_games(players_mdf_2016_filled)
games_2017 = summarize_games(players_mdf_2017_filled)
games_2018 = summarize_games(players_mdf_2018_filled)
```

```python
games_2016["season"] = "2016-17"
games_2017["season"] = "2017-18"
games_2018["season"] = "2018-19"
```

```python
combined_df = pd.concat([games_2016, games_2017, games_2018], ignore_index=True)
```
